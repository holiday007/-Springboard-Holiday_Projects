```python
# preparing the environment
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

plt.rcParams['figure.figsize'] = [12, 7]
```

```python
# read data
df = pd.read_csv('ks-projects-201801-ready.csv')
df = df.drop(['Unnamed: 0'], axis=1) # the first column is unnamed and repetitive of the index column
df.head()
```

# Visual EDA

## Overall of project success rate

```python
# data frame for percentage of projects in different states
success_rate = df[['state', 'ID']].groupby('state').count()
success_rate.columns = ['count']
success_rate['percentage'] = success_rate['count'] * 100 / success_rate['count'].sum()

success_rate = success_rate.drop('count', axis = 1)

# visualize
plt.figure(figsize = (12,7))
sns.barplot(x=success_rate.index, y='percentage', data=success_rate)
plt.xticks(rotation=45)
plt.show()
```

As we can see, about 10% of the projects were canceled, over 50% of them failed, and about 36% of them were successful

## Projects across time

### Number of projects across years

```python
# create a data_frame of launch years and counts
launch_yearcount = df[['launch_year', 'ID']].groupby('launch_year').count()
launch_yearcount.columns = ['count']
# create a data_frame of deadline years and counts
deadline_yearcount = df[['deadline_year', 'ID']].groupby('deadline_year').count()
deadline_yearcount.columns = ['count']

# visualize the counts of projects over time
count_launcyear = launch_yearcount.plot(kind='bar', color = 'blue')
plt.xlabel('Launch Year')
plt.ylabel('Count')
plt.legend(['Count of projects'], loc='upper left')
plt.show()

count_deadlineyear = deadline_yearcount.plot(kind='bar', color = 'blue')
plt.xlabel('Deadline Year')
plt.ylabel('Count')
plt.legend(['Count of projects'], loc='upper left')
plt.show()
```

Here are a couple important observations:
1. Not many projects were seen launched or due in 2018, this is probabaly because the data was released in 2018-01, so it didn't include many 2018 projects
2. 2015 had the most projects launched or due
3. There is a dip of considerable size from 2015 to 2016. 
4. Kickstarted was founded in **2009**, yet oddly there were projects due 1970, some sort of mistake must have taken place

### Success Rate across years

```python
success_rate={}

for year in range(2009, 2019):
    successful = len(df[(df.deadline_year==year) & (df.state=='successful')])
    total = len(df[df.deadline_year==year])
    success_rate[year] = successful/total

df_sr = pd.DataFrame.from_dict(success_rate, orient='index', columns=['success_ratio'])
print(df_sr)

# visualize success rate on a bar plot
df_sr.plot(kind='bar', color = 'blue')
plt.xlabel('Year of deadline')
plt.ylabel('Ratio')
plt.legend(['success ratio'], loc = 'upper right')
plt.show()
```

There are a couple interesting observations here:
1. from 2009 to 2013, the success rate seems overall stable with a value slightly above 0.4, and a peak at 2011 of 0.47
2. There was a dip of success rate of considerable size from 2013 to 2014
3. 2015 saw the lowest success rate out of all. Interestingly, 2015 also witnessed the highest number of projects launched
    - the all-time low success rate (excluding 2018) in 2015 may have discouraged people from launching new projects, thus resulted in the dip in number of projects from 2015 to 2016
4. 2018 witnesses the lowest success rate, 0.19, for the past decade, this number is in particular low, let's inspect this further

```python
# let's look at the project state of 2018
df[(df.deadline_year == 2018)].state.value_counts()
```

Since only projects in January were included in 2018, maybe this low success rate is a product of seasonality? Let's inspect the success rate by months to test this assumption

```python
# create a data frame of success rate by month
success_rate_month={}

for month in range(1, 13):
    successful = len(df[(df.deadline_month==month) & (df.state=='successful')])
    total = len(df[df.deadline_month==month])
    success_rate_month[month] = successful/total

df_monthly_sr = pd.DataFrame.from_dict(success_rate_month, orient='index', columns=['success_ratio'])
print(df_monthly_sr)

# visualiza success rate by month
df_monthly_sr.plot(kind='bar', legend=None, color = 'blue')
plt.xlabel('Month')
plt.title('Success rate across months bar plot')
plt.show()

df_monthly_sr.plot(legend=None, color = 'blue')
plt.xlabel('Month')
plt.title('Success rate across months line plot')
plt.show()
```

Seasonality does indicate that **success rate is lowest in January**. However, the overall January success rate is around 0.33, which is still much higher than the success rate in 2018, 0.19. Thus, there is something more than just seasonality at work. 

One assumption is that the low-quality projects are likely to be failed in an early stage, meanwhile the ones with better potential are still being evaluated, which means that they are still alive, which is why they are not seen in this data set where only projects that have reached a conclusive state (successful, failed, canceled) were included.

### Success rate across main categories
Now let's take a look at the success rate across main categories

```python
# data frame for success rate by category
success_maincat = {}

unique_maincat = list(df.main_category.unique())

for i in range(len(df.main_category.unique())):
    successful = len(df[(df.main_category == unique_maincat[i]) & (df.state=='successful')])
    
    total = len(df[df.main_category==unique_maincat[i]])
    
    success_maincat[unique_maincat[i]] = successful/total

df_sr_maincat = pd.DataFrame.from_dict(success_maincat, orient = 'index', columns=['success_ratio'])

df_sr_maincat = df_sr_maincat.sort_values(by = 'success_ratio', ascending=False)

# plot in bar plot
plt.figure(figsize=(12,7))
countplot = sns.barplot(x=df_sr_maincat.index, y='success_ratio', data = df_sr_maincat)
plt.xticks(rotation=60)
plt.show()
```

We can see from the success ratio that:
- The top 3 main categories that are most likely to succeed are "Dance", "Theater" and "Comics"
- The 3 main categories that are least likely to succeed are "crafts", "Journalism" and "Technology" 

It seems like main category is related to a project's success chance

## Projects across categories
Now let's take a peek of the projects across categories, since the amount of categories is large, we are only going to look at the 20 categories with the most project counts. Maybe we can still gain some insights from this subset 

### Project counts across categories

```python
# Data frame
count_cat = df[['category', 'ID']].groupby('category').count()
count_cat.columns = ['count']
count_cat = count_cat.sort_values(by='count', ascending=False)[:20]

print("The first 20 categories with the most projects account for", round(count_cat['count'].sum() * 100/len(df)), "% of all projects")
```

```python
# Barplot
plt.figure(figsize=(12,7))
sns.barplot(x=count_cat.index, y='count', data=count_cat)
plt.xticks(rotation=60)
plt.title('Categories with top 20 counts of projects')
plt.show()
```

### Success rate across these categories

```python
# Data frame for the success rate across these 20 categories

# subset these categories out from original df
df_top20cat = df[(df.category.isin(list(count_cat.index)))]
sr_cat = {}
cat_list = list(count_cat.index)

for i in range(20):
    successful = len(df_top20cat[(df_top20cat.category == cat_list[i]) & (df_top20cat.state == 'successful')])
    total = len(df_top20cat[df_top20cat.category == cat_list[i]])
    success_rate = successful/total
    
    sr_cat[cat_list[i]] = success_rate

sr_cat = pd.DataFrame.from_dict(sr_cat, orient='index', columns=['success_rate'])
sr_cat = sr_cat.sort_values(by='success_rate', ascending=False)

# Visualize the data
plt.figure(figsize=(12,7))
sns.barplot(x=sr_cat.index, y='success_rate', data=sr_cat)
plt.xticks(rotation=60)
plt.show()
```

As we can see, different categories can have very different success rate, and there are an interesting observation worthy of mentioning:
1. "Tablettop Games" have a success chance above 0.6, and it's main category, "Games" only has a success chance slightly below 0.4, this indicates that category might reveal some subtle variations main category may have failed to capture
